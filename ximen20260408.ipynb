````python
import json
import urllib.request as req

url = "https://api.gamer.com.tw/anime/v1/danmu.php?videoSn=36632&geo=TW%2CHK"
f = req.urlopen(url)
content = f.read()
# 型態轉換: bytes => list/dict(JSON)
print(type(content))
content = json.loads(content)
print(type(content))
````

````python
# 一群學生: [學生0, 學生1, 學生2]
# 字典(dict): 聚集很多欄位來表示一個複雜單位
# {}: key(int/str):value (欄位名稱:直)
# 無序集合
# 順序: key=0,1,2,3
student = {
    "sn":0,
    "name":"Elwing",
    "height":175
}
print(student)
# 查詢: key  東西[key]
print(student["name"])
student["weight"] = 80
print(student)
````
