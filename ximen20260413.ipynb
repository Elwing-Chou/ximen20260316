````python
# 完整版抓取
import json
import urllib.request as req
import pandas as pd
import os

# BeautifulSoup
import bs4 as bs

url = "https://ani.gamer.com.tw/animeVideo.php?sn=36632"
h = {
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/147.0.0.0 Safari/537.36"
}
r = req.Request(url,headers=h)
f = req.urlopen(r)
content = f.read()
# 之前的json.loads
html = bs.BeautifulSoup(content)
# 大區塊.find/find_all
for a in html.find_all("a"):
    # 目標區快: 1. 拿特別屬性 ["href"] 2. 拿顯示文字: .get_text()
    a_href = a["href"]
    a_text = a.get_text()
    # 字串的專屬功能: google "Python string method"
    if a_href.startswith("?sn="):
        print(a_href)
        # 用=切開
        sn = a_href.split("=")[1]
        # 組合上面
        dn = "baha_data"
        if not os.path.exists(dn):
            os.makedirs(dn)
        url = "https://api.gamer.com.tw/anime/v1/danmu.php?videoSn=" + sn + "&geo=TW%2CHK"
        f = req.urlopen(url)
        content = f.read()
        content = json.loads(content)
        danmu = content["data"]["danmu"]

        df = pd.DataFrame(danmu)
        # 額外增加集數欄位(把所有格子設成同一個職)
        df["chapter"] = a_text
        fp = dn + "/" + a_text + ".csv"
        df.to_csv(fp, encoding="utf-8")
````

````python
# 完整版的發文id分析
import glob

# 如果資料夾不存在, 你就開起來
dn = "baha_2"
if not os.path.exists(dn):
    os.makedirs(dn)

# go through每一個csv
tables = []
for fn in glob.glob("baha_data/*.csv"):
    partial = pd.read_csv(fn, encoding="utf-8", index_col=0)
    # 把東西方到list李 append
    tables.append(partial)

# 把list的所有df合併再一起
df = pd.concat(tables)
# value_counts/unique
for u in df["userid"].unique():
    fil = df["userid"] == u
    fn = dn + "/" + u + ".csv"
    df[fil].to_csv(fn, encoding="utf-8")
````
