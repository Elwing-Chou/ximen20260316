````python
import json
import urllib.request as req
import pandas as pd

url = "https://api.gamer.com.tw/anime/v1/danmu.php?videoSn=36632&geo=TW%2CHK"
f = req.urlopen(url)
content = f.read()
content = json.loads(content)
danmu = content["data"]["danmu"]

df = pd.DataFrame(danmu)
df.to_csv("danmu.csv", encoding="utf-8")
df
````

````python
import os
import pandas as pd

# 如果資料夾不存在, 你就開起來
dn = "baha"
if not os.path.exists(dn):
    os.makedirs(dn)

df = pd.read_csv("danmu.csv", encoding="utf-8", index_col=0)
# value_counts/unique
for u in df["userid"].unique():
    fil = df["userid"] == u
    fn = dn + "/" + u + ".csv"
    df[fil].to_csv(fn, encoding="utf-8")
````
