````python
# 讓使用者輸入身高體重來計算bmi
# bmi = 體重(kg) / 身高(m) ^ 2
weight = float(input("請輸入體重:"))
height = float(input("請輸入身高:"))
bmi = weight / (height / 100) ** 2
print("體重是:" + str(weight))
print("身高是:" + str(height))
print("bmi是:" + str(bmi))

if bmi > 25:
    print("過重")
else:
    print("正常")
````
