````python
import random

trans = ["剪刀", "石頭", "布"]

com = random.randint(0, 2)
my = int(input("0.剪刀 1.石頭 2.布:"))

print("我出的:" + trans[my])
print("電腦的:" + trans[com])

if my == (com + 1) % 3:
    print("WIN")
elif com == (my + 1) % 3:
    print("LOSE")
else:
    print("EVEN")
````
