````python
import random

trans = ["剪刀", "石頭", "布"]

com = random.randint(0, 2)
my = int(input("0.剪刀 1.石頭 2.布:"))

print("我出的:" + trans[my])
print("電腦的:" + trans[com])

# 我出的拳是電腦的下一個
if my == (com + 1) % 3:
    print("WIN")
elif my == (com - 1) % 3:
    print("LOSE")
else:
    print("EVEN")
````

````python
# //: 取商 %: 取餘
import random

trans = ["剪刀(0)", "石頭(0)", "布(0)",
         "剪刀(1)", "石頭(1)", "布(1)"]

# com = random.randint(0, 5)
com = 3
my = int(input("0-5:"))

print("我出的:" + trans[my])
print("電腦的:" + trans[com])

# //: 同組
if com // 3 == my // 3:
    print("same group")
    # case1:0-2 case2:3-5
    my2 = my % 3
    com2 = com % 3
    # my2, com2: 0-2
    # shift(反轉)+tab/tab
    if my2 == (com2 + 1) % 3:
        print("WIN")
    elif my2 == (com2 - 1) % 3:
        print("LOSE")
    else:
        print("EVEN")
else:
    print("EVEN(dif group)")
````
