```python
# list[key]
scores = [80, 60, 70, 50]

count = 0
i = 0
while i < len(scores):
    s = scores[i]
    if s >= 60:
        count = count + 1
    i = i + 1
print(count)
```

```python
# 上面這個寫法的侷限性:
# 以後會遇到某些群集 key不是0 1 2 3 scores[i]就無法用
# 如果迴圈: 是走過群集 for in (走過) 群集(list/str)
# 保證群集裡的每個東西都會拿出來給你一遍
# 對於有序(key=0 1 2 3) 就會照順序給你
scores = [80, 60, 70, 50]

count = 0
for s in scores:
    if s >= 60:
        count = count + 1
print(count)
```

```python
# 目前累加的總和
total = 0
i = 0
while i < 10:
    total = total + (i + 1)
    print(total)
    i = i + 1
```

```python
total = 0
for i in [0, 1, 2, 3, 4]:
    total = total + (i + 1)
    print(total)

# range(5) -> [0, 1, 2, 3, 4] (o)
# range(2, 5) -> [2, 3, 4]
# range(1, 7, 2) -> [1, 3, 5]
total = 0
for i in range(5):
    total = total + (i + 1)
    print(total)
```

```python
# https://zh.wikipedia.org/zh-tw/E_(%E6%95%B0%E5%AD%A6%E5%B8%B8%E6%95%B0)
# 1 / 0! + 1 / 1! + 1 / 2!
total = 0
frac = 1

i = 0
while i < 100:
    # 0! 1! 2! 3!
    if not i == 0:
        frac = frac * i
    total = total + (1 / frac)
    i = i + 1
print(total)
```
