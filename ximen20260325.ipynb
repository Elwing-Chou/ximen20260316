```python
# list[key]
scores = [80, 60, 70, 50]

count = 0
i = 0
while i < len(scores):
    s = scores[i]
    if s >= 60:
        count = count + 1
    i = i + 1
print(count)
```

```python
# 上面這個寫法的侷限性:
# 以後會遇到某些群集 key不是0 1 2 3 scores[i]就無法用
# 如果迴圈: 是走過群集 for in (走過) 群集(list/str)
# 保證群集裡的每個東西都會拿出來給你一遍
# 對於有序(key=0 1 2 3) 就會照順序給你
scores = [80, 60, 70, 50]

count = 0
for s in scores:
    if s >= 60:
        count = count + 1
print(count)
```
