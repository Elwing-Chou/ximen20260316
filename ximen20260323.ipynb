````python
a, b = 33, 5
result = str(a // b) + "."
# 37 % 7 -> 2
a = a % b
i = 0
while i < 10:
    # 2 -> 20
    a10 = a * 10
    # 20 // 7 -> 2   0.2
    result = result + str(a10 // b)
    # 20 % 7
    a = a10 % b
    print(result)
    if a == 0:
        print("整除")
        break
    i = i + 1
````
