````python
i = 0
while i < 10:
    print(i, i+1, 50-i, 2*i+1)
    i = i + 1
````

````python
a, b = 33, 5
result = str(a // b) + "."
# 37 % 7 -> 2
a = a % b
i = 0
while i < 10:
    # 2 -> 20
    a10 = a * 10
    # 20 // 7 -> 2   0.2
    result = result + str(a10 // b)
    # 20 % 7
    a = a10 % b
    print(result)
    if a == 0:
        print("整除")
        break
    i = i + 1
````

````python
# o
# oo
# ooo
# oooo
# ooooo
i = 0
while i < 5:
    print("o" * (i + 1))
    i = i + 1
````

````python
#   o
#  oo
# ooo
lsymbol, rsymbol = " ", "o"
layer = 10
i = 0
while i < layer:
    lcount = layer - i - 1
    rcount = i + 1
    line = (lsymbol * lcount +
            rsymbol * rcount)
    print(line)
    i = i + 1
````

````python
# xxxxxxx
#  ooooo
#   xxx
#    o

layer = 10
i = 0
while i < layer:
    if i % 2 == 0:
        symbol1 = "_"
        symbol2 = "x"
        symbol3 = "_"
    else:
        symbol1 = "_"
        symbol2 = "o"
        symbol3 = "_"

    count1 = i
    count2 = (2 * layer - 1) - 2 * i
    count3 = i
    line = (symbol1 * count1 +
            symbol2 * count2 +
            symbol3 * count3)
    print(line)
    i = i + 1
````

````python
# extra. 供參考
symbol1_list = ["_"]
symbol2_list = ["o", "x", "g"]
symbol3_list = ["_"]

layer = 10
i = 0
while i < layer:
    symbol1 = symbol1_list[i % len(symbol1_list)]
    symbol2 = symbol2_list[i % len(symbol2_list)]
    symbol3 = symbol3_list[i % len(symbol3_list)]

    count1 = i
    count2 = (2 * layer - 1) - 2 * i
    count3 = i
    line = (symbol1 * count1 +
            symbol2 * count2 +
            symbol3 * count3)
    print(line)
    i = i + 1
````
