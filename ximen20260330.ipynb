```python
import ColabTurtle.Turtle as turtle

turtle.initializeTurtle(initial_speed=4, initial_window_size=(800,500))

many = 5
d = 100
angle = 360 / many

# 紀錄原點
x, y = turtle.pos()
# 往前100
turtle.forward(d)
# 紀錄等一下畫回來的點
x1, y1 = turtle.pos()
# 回到原點
turtle.goto(x, y)
turtle.right(angle)
turtle.forward(d)
# 回到剛剛的紀錄點
turtle.goto(x1, y1)
turtle.goto(x, y)
```

```python
# https://github.com/Elwing-Chou/ximen20260316
# 快捷鍵: TAB/Shift+TAB
import random
import ColabTurtle.Turtle as turtle

turtle.initializeTurtle(initial_speed=10, initial_window_size=(800,500))

many = 100
d = 100
angle = 360 / many

i = 0
while i < many:
    r = random.randint(0, 255)
    g = random.randint(0, 255)
    b = random.randint(0, 255)
    turtle.color(r, g, b)

    x, y = turtle.pos()
    turtle.forward(d)
    x1, y1 = turtle.pos()
    turtle.goto(x, y)
    turtle.right(angle)
    turtle.forward(d)
    turtle.goto(x1, y1)
    turtle.goto(x, y)
    i = i + 1
```

```python
import random

trans = ["L", "E", "W"]
win, even, lose = 0, 0, 0
i = 0
while i < 10:
    # -1: lose 0: even 1: win
    result = random.randint(-1, 1)
    print(trans[result+1])
    if result == -1:
        lose = lose + 1
    elif result == 0:
        even = even + 1
    elif result == 1:
        win = win + 1
    i = i + 1
print(lose, even, win)
```
